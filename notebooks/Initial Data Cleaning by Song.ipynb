## Make "SEX" Column Categorical
1 = male; 2 = female 
<br>
We consider values in "SEX" column as categorical. We create two columns "MALE" and "FEMALE" with values {True,False} to store the values in "SEX".

```python
data["MALE"]=pd.get_dummies(data["SEX"])[1]
data["FEMALE"]=pd.get_dummies(data["SEX"])[2]
```

## Rename Column "PAY_0" as "PAY_1"

```python
data.rename(columns={"PAY_0":"PAY_1"})
```

## Modify the "EDUCATION" Column
1 = graduate school; 2 = university; 3 = high school; 0, 4, 5, 6 = others
<br>The values "0,4,5,6" in the original data all refer to "others". We merge these cases as "0".

```python
data["EDUCATION"].value_counts()
```

```python
def eduModify(n):
    if n>3 or n==0: return 0
    else: return n
```

```python
data["EDUCATION"]=data["EDUCATION"].apply(eduModify)
```

```python
data["EDUCATION"].value_counts()
```
